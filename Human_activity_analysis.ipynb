# 3. Supervised Learning

## 3.1 Ensemble Learning: Random Forest Classifier (Averaging Method- 1)

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
```

### 3.1.1 Training the classifier without feature selection

```python
X_train=data_train.drop('Activity',axis=1)
Y_train=data_train['Activity']
```

```python
X_test=data_test.drop('Activity',axis=1)
Y_test=data_test['Activity']
```

```python
import time
start=time.time()
rf=RandomForestClassifier(n_estimators=200,random_state=0)
rf=rf.fit(X_train,Y_train)
rf_predictions=rf.predict(X_test)
elapsed=time.time()-start
```

```python
print('The Accuracy of the Random Forest Classifier on the Testing set: ',accuracy_score(Y_test,rf_predictions))
```

```python
print('The running time of the RF classifier: '+str(elapsed)+' [s]')
```

```python
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

x_cols=list(X_train.columns)
# Print the feature ranking
print("Feature ranking:")

for f in range(X_train.shape[1]):
    print("%d. feature %s : %f" % (f + 1, x_cols[f], importances[indices[f]]))
```

```python
model = SelectFromModel(rf, prefit=True)
X_train_red_rf_features = model.transform(X_train)
print(X_train_red_rf_features.shape)
```

```python
X_test_red_rf_features=model.transform(X_test)
print(X_test_red_rf_features.shape)
```

### 3.1.2 Training the classifier with feature selection

```python
import time
start=time.time()
rf_red=RandomForestClassifier(n_estimators=200,random_state=0)
rf_red=rf_red.fit(X_train_red_rf_features,Y_train)
rf_predictions_red=rf_red.predict(X_test_red_rf_features)
elapsed=time.time()-start
```

```python
print('The Accuracy of the Random Forest Classifier on the Testing set: ',accuracy_score(Y_test,rf_predictions_red))
```

```python
print('The running time of the RF classifier: '+str(elapsed)+' [s]')
```

## 3.2 Ensemble Learning: Extra Trees Classifier (Averaging Method- 2)

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
import time
start=time.time()
etf=ExtraTreesClassifier(random_state=0)
etf=etf.fit(X_train,Y_train)
etf_predictions=etf.predict(X_test)
elapsed=time.time()-start
```

```python
print('The Accuracy of the Extra Trees Classifier on the Testing set: ',accuracy_score(Y_test,etf_predictions))
```

```python
print('The running time of the EtC classifier: '+str(elapsed)+' [s]')
```

```python
importances = etf.feature_importances_
indices = np.argsort(importances)[::-1]

x_cols=list(X_train.columns)
# Print the feature ranking
print("Feature ranking:")

for f in range(X_train.shape[1]):
    print("%d. feature %s : %f" % (f + 1, x_cols[f], importances[indices[f]]))
```

```python
model = SelectFromModel(etf, prefit=True)
X_train_red_etf_features = model.transform(X_train)
print(X_train_red_etf_features.shape)
```

```python
X_test_red_etf_features=model.transform(X_test)
print(X_test_red_etf_features.shape)
```

```python
import time
start=time.time()
etf_red=ExtraTreesClassifier(random_state=0)
etf_red=etf_red.fit(X_train_red_etf_features,Y_train)
etf_predictions_red=etf_red.predict(X_test_red_etf_features)
elapsed=time.time()-start
```

```python
print('The Accuracy of the Extra Trees Classifier on the Testing set: ',accuracy_score(Y_test,etf_predictions_red))
```

```python
print('The running time of the EtC classifier: '+str(elapsed)+' [s]')
```
